Mencari Outlayer dari setiap data

```python
# Select only numerical columns for outlier detection
numerical_df = customers_df.select_dtypes(include=np.number)

q25, q75 = np.percentile(numerical_df, 25), np.percentile(numerical_df, 75)
iqr = q75 - q25
cut_off = iqr * 1.5
minimum, maximum = q25 - cut_off, q75 + cut_off

# Calculate outliers for each numerical column
outliers_customers = []
for col in numerical_df.columns:
  col_outliers = [x for x in numerical_df[col] if x < minimum or x > maximum]
  outliers_customers.extend(col_outliers)

print("Jumlah outliers", len(outliers_customers))
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kota mana yang memiliki jumlah pelanggan terbanyak?

```python
sum_city_df = customers_df.groupby("customer_city").customer_city.count().sort_values(ascending=False).reset_index(name='count')
sum_city_df.head(15)
```

```python
fig, ax = plt.subplots(figsize=(24, 6))

# Menghitung jumlah customer per kota
city_counts = customers_df['customer_city'].value_counts().reset_index()
city_counts.columns = ['customer_city', 'customer_count']

colors = ["#3A6D8C", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Plot bar chart untuk kota dengan customer terbanyak
sns.barplot(x="customer_count", y="customer_city", data=city_counts.head(5), palette=colors, ax=ax)
ax.set_ylabel("Name of City")
ax.set_xlabel("Number of Customers")
ax.set_title("Top 5 Cities with Most Customers", loc="center", fontsize=15)
ax.tick_params(axis='y', labelsize=12)

plt.suptitle("Cities by Number of Customers", fontsize=20)
plt.show()
```

### Pertanyaan 3: Bagaimana perbandingan jumlah pesanan dengan status *Delivered* pada tahun 2016 hingga 2018?

```python
# Filter data hanya untuk status 'delivered'
delivered_orders_df = orders_df[orders_df['order_status'] == 'delivered']

monthly_delivered_df = delivered_orders_df.resample(rule='M', on='order_purchase_timestamp').agg({
    "order_id": "nunique"  # Menghitung jumlah pesanan unik yang delivered per bulan
}).rename(columns={"order_id": "delivered_count"})

# Ubah index menjadi format 'YYYY-MM' untuk menampilkan tahun-bulan
monthly_delivered_df.index = monthly_delivered_df.index.strftime('%Y-%m')

# Reset index agar 'order_date' menjadi kolom
monthly_delivered_df = monthly_delivered_df.reset_index().rename(columns={"order_purchase_timestamp": "order_date"})

# Menampilkan hasil data pesanan delivered dari tahun ke tahun dan bulan ke bulan
monthly_delivered_df
```

```python
# Filter data untuk status delivered dari seluruh tahun yang ada
delivered_orders_df = orders_df[orders_df['order_status'] == 'delivered']

# Membuat kolom 'year' dan 'month' untuk memisahkan informasi tahun dan bulan
delivered_orders_df['year'] = delivered_orders_df['order_purchase_timestamp'].dt.year
delivered_orders_df['month'] = delivered_orders_df['order_purchase_timestamp'].dt.month

# Mengelompokkan data berdasarkan tahun dan bulan
monthly_delivered_df = delivered_orders_df.groupby(['year', 'month']).agg({
    "order_id": "nunique"  # Menghitung jumlah pesanan unik yang delivered
}).rename(columns={"order_id": "delivered_count"}).reset_index()

# Plot line chart untuk jumlah 'delivered' per bulan, dengan warna berbeda untuk setiap tahun
plt.figure(figsize=(12, 6))

# Daftar warna yang kontras untuk digunakan pada setiap tahun
colors = plt.cm.get_cmap('tab20', len(monthly_delivered_df['year'].unique()))

# Loop untuk memplot setiap tahun dengan warna berbeda
for i, year in enumerate(monthly_delivered_df['year'].unique()):
    year_data = monthly_delivered_df[monthly_delivered_df['year'] == year]
    plt.plot(year_data['month'], year_data['delivered_count'], label=str(year), color=colors(i), marker='o', linewidth=2)

# Title and labels
plt.title("Total Orders by Month (2016-2018", fontsize=20)
plt.xlabel("Month", fontsize=15)
plt.ylabel("Number of Delivered Orders", fontsize=15)

# Menampilkan label bulan (1-12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=10)

# Menambahkan legenda untuk membedakan setiap tahun
plt.legend(title="Year", fontsize=12)

plt.tight_layout()

# Menampilkan plot
plt.show()
```
